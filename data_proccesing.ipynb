```python
row_removal = ["Coming Soon", 'Mobile house for sale']
df1 = df1[~df1['TYPE'].isin(row_removal)]
df1['TYPE'] = df1['TYPE'].replace('Condop for sale', 'Condo for sale')
df1['TYPE'] = df1['TYPE'].apply(lambda x: x.split(' ')[0])
df2 = df1
df2.groupby('TYPE')['TYPE'].agg('count')
```

```python
df4 = df3.copy()
df4['PRICEPERSQFT'] = df4['PRICE'] / df['PROPERTYSQFT']
df4 = df4.apply(lambda x: x.astype(int) if x.dtype == 'float64' else x)
df4
```

```python
sublocalities_to_replace = ["Snyder Avenue", "Brooklyn Heights", "Coney Island", "Dumbo", "Fort Hamilton"]

# Replace the specified sublocalities with 'Brooklyn'
df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(sublocalities_to_replace, 'Brooklyn')

# Group by SUBLOCALITY and count occurrences
df4.groupby("SUBLOCALITY")['SUBLOCALITY'].agg('count')
```

```python
replace = ["Flushing", "Jackson Heights", "Rego Park"]

df4['SUBLOCALITY'] = df4["SUBLOCALITY"].replace(replace, "Queens")

df4.groupby('SUBLOCALITY')['SUBLOCALITY'].agg('count')
```

```python
replace = ["East Bronx", "Riverdale"]

df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(replace, "Bronx County")

df4.groupby("SUBLOCALITY")['SUBLOCALITY'].agg('count')
```

```python
counties = df4[df4['SUBLOCALITY'].str.contains('County')]

list_of_counties = counties['SUBLOCALITY'].unique().tolist()

for i in list_of_counties:
    if i == 'New York County':
        df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(i, 'Manhattan')
    elif i == 'Richmond County':
        df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(i, 'Staten Island')
    elif i == "Kings County":
        df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(i, 'Brooklyn')
    else:
        df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace(i, i.split(' ')[0])


df4['SUBLOCALITY'] = df4['SUBLOCALITY'].replace('The Bronx', 'Bronx')

df4.groupby('SUBLOCALITY')['SUBLOCALITY'].agg('count')
```

```python
#removes any outliers for square footage, usual square footage is 300 sqft per bedroom
df5 = df4[~(df4.PROPERTYSQFT / df4.BEDS < 300)]
df5
```

```python
#removes any extreme prices that might skew our model
def standard_deivation(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('SUBLOCALITY'):
        m = np.mean(subdf.PRICEPERSQFT)
        st = np.std(subdf.PRICEPERSQFT)
        reduced_df = subdf[(subdf.PRICEPERSQFT > (m -st)) & (subdf.PRICEPERSQFT <= (m + st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    
    return df_out
```

```python
df6 = standard_deivation(df5)
df6
```

```python
def scatter_plot(df, location):
    bed_2 = df[(df.SUBLOCALITY==location) & (df.BEDS==1)]
    bed_3 = df[(df.SUBLOCALITY==location) & (df.BEDS==2)]
    
    plt.scatter(bed_2.PROPERTYSQFT, bed_2.PRICE, color='blue', label='2 Beds', s=50)
    plt.scatter(bed_3.PROPERTYSQFT, bed_3.PRICE, color='red', marker='+', label='3 Beds', s=50)

    plt.xlabel('Total Square Feet')
    plt.ylabel('Price Per SQFT')

    plt.title(location)
    plt.show()
```

```python
locations = df6['SUBLOCALITY'].unique().tolist()


for i in locations:
    scatter_plot(df6, i)
```

```python
# removes houses in the same location in which a lower amount of beds is more expensive
def remove_bed_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('SUBLOCALITY'):
        bed_info = {}
        for bed, bed_df in location_df.groupby('BEDS'):
            bed_info[bed] = {
                'mean': np.mean(bed_df.PRICEPERSQFT),
                'std': np.std(bed_df.PRICEPERSQFT),
                'count': bed_df.shape[0]
            }
        
        for bed, bed_df in location_df.groupby('BEDS'):
            info = bed_info.get(int(bed) - 1)
            if info and info['count'] > 5:
                exclude_indices = np.append(exclude_indices, bed_df[bed_df.PRICEPERSQFT < (info['mean'])].index.values)
    
        return df.drop(exclude_indices, axis='index')
        
```

```python
df7 = remove_bed_outliers(df6)

df7
```

```python
# futher outlier removal
df8 = df7[df7.BATH < df7.BEDS+2]

df8
```

```python
# one hot encoding
dummies = pd.get_dummies(df8.SUBLOCALITY, dtype=int)

df9 = pd.concat([df8, dummies], axis='columns')

df9 = df9.drop('SUBLOCALITY', axis='columns')

df9
```

```python
# one hot encoding
dummies = pd.get_dummies(df9.TYPE, dtype=int)

df10 = pd.concat([df9, dummies], axis='columns')

df10 = df10.drop('TYPE', axis='columns')

df10
```

```python
# ml model training

X = df10.drop('PRICE', axis='columns')
y = df10.PRICE

# train/test split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=10)
```

```python
from sklearn.tree import DecisionTreeRegressor
import joblib
model = DecisionTreeRegressor(random_state=0)
model.fit(X_train, y_train)

model.score(X_test, y_test)

joblib.dump(model, 'nyc_housing_price_model.joblib')
```

```python
loaded_model = joblib.load('nyc_housing_price_model.joblib')

prediction = loaded_model.predict(X_test)

df = pd.DataFrame({'Real Values':y_test, 'Predicted Values':prediction})

df
```
